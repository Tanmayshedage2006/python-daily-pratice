# BREAK

```python
for i in range(1,6):
    if i == 3:
        break
    else:                  # just stop the excuetion of the programm
        print(i)
        print('Hello')
```

# CONTINUE

```python
for i in range(1,6):
    if i == 3:
        continue           # This says that go to the next iteration
    else:
        print(i)
     

```
