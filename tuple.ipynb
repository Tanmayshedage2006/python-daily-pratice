```python
# Craeating a Tuple
my_tuple=(123)
print(my_tuple)
```

```python
# nested tuple
a=("data",[123],(986))
print(a)
type(a)
```

```python
# indexing on a Tuple
a=(47,48,90,44)
print(a[3])
```

```python
# Slicing index
a=(12,34,24,23,34)
print(a[0:2])
print(a[::1])
print(a[::-1])
```
