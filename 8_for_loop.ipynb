```python
total= int (input('Total bill'))
if total > 100 :
    discount =  total*0.2
    final_price=total-discount
    print(f'Total was {total} and we have given you {discount} and final price will be {final_price}')
elif total > 50 :
    discount =  total*0.1
    final_price=total-discount
    print(f'Total was {total} and we have given you {discount} and final price will be {final_price}')
else :
    print(f'Total was {total} and you dont have any discount')
```
